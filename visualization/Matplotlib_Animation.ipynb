```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
```

```
plt.rc('animation', html='html5')
```

```
def func(x): return 1/(1+np.exp(x)) * np.sin(x)
```

```
sz = 100
```

```
x = 2 * ((np.pi - (-np.pi)) * np.random.random_sample(size=sz) + (-np.pi))
```

```
y = func(x) + 0.2 * np.random.random_sample(size=sz)
```

```
plt.scatter(x, y)
```

```
epochs = 20

fig = plt.figure(figsize=(14, 8))
ax = fig.add_subplot(1, 1, 1)
ax.scatter(x, y, c='orange', label='Ground truth')
ax.set_ylim(-1, 6)
ax.set_title('Ground truth & Model', fontsize=16)
line, = ax.plot([], [], label='Model')
ax.legend(loc='lower right')
title = fig.suptitle(f'Epoch: 0/{epochs}', fontsize=22)
plt.close()

def animate(i):
    xi = x[]
    line.set_data(xi, y * max(0, 10-i)*np.random.rand())
    title.set_text(f'Epoch: {i+1}/{epochs}')
    return line, title
```

```
animation.FuncAnimation(fig, animate, range(epochs), interval=60)
```
