## Helpers

```
import tensorflow as tf
import tensorflow_probability as tfp
import matplotlib.pyplot as plt
```

```
tf.random.set_seed(31)
```

```
def plot_distribution(samples):
  plt.figure(figsize=(10, 6))
  # Show 4 different binwidths
  for i, binwidth in enumerate([10, 50, 100, 150]):

    # Set up the plot
    ax = plt.subplot(2, 2, i + 1)

    # Draw the plot
    ax.hist(samples, bins = int(len(samples)/binwidth), color = 'blue', edgecolor = 'black')

    # Title and labels
    ax.set_title('Histogram with Binwidth = %d' % binwidth, size = 12)
    ax.set_xlabel('Sample', size = 10)
    ax.set_ylabel('Count', size= 10)

  plt.tight_layout()
  plt.show()
```

## Continous probability distributions

### Uniform

```
samples = tf.random.uniform([1000], minval=0.0, maxval=1.0)
plot_distribution(samples)
```

### Normal

```
samples = tf.random.normal([1000], mean=0.0, stddev=1.0)
plot_distribution(samples)
```

## Discrete probability distributions

### Poisson

```
samples = tf.random.poisson([1000], lam=2)
plot_distribution(samples)
```

### Binomial

```
samples = tf.keras.backend.random_binomial([1000], p=0.2)
plot_distribution(samples)
```
